```python
# ====================================================================================
# 1. ë¼ì´ë¸ŒëŸ¬ë¦¬ Import
# ====================================================================================
import os
import random
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from sklearn.model_selection import train_test_split
from tqdm.notebook import tqdm

# --- ë¹„êµ ì‹¤í—˜ì„ ìœ„í•œ í† í¬ë‚˜ì´ì € ---
import sentencepiece as spm
from konlpy.tag import Mecab

# ====================================================================================
# 2. ì¤‘ì•™ ì„¤ì • (Configuration)
# ====================================================================================
CFG = {
    # --- ì‹¤í—˜ ì„ íƒ ---
    'TOKENIZER_TYPE': 'KoNLPy', # 'SentencePiece' ë˜ëŠ” 'KoNLPy' ì„ íƒ
    'KONLPY_TOKENIZER': 'Mecab',     # TOKENIZER_TYPEì´ 'KoNLPy'ì¼ ê²½ìš°, ì‚¬ìš©í•  í˜•íƒœì†Œ ë¶„ì„ê¸°
    
    # --- SentencePiece í•˜ì´í¼íŒŒë¼ë¯¸í„° ---
    'SP_VOCAB_SIZE': 10000,          # SentencePiece ë‹¨ì–´ ì§‘í•© í¬ê¸°
    'SP_MODEL_TYPE': 'bpe',          # 'bpe' ë˜ëŠ” 'unigram'
    
    # --- KoNLPy í•˜ì´í¼íŒŒë¼ë¯¸í„° (ì¶”ê°€) ---
    'KONLPY_VOCAB_SIZE': 10000,      # KoNLPy ì‚¬ìš© ì‹œ ë‹¨ì–´ ì§‘í•© í¬ê¸° ì œí•œ
    
    # --- ëª¨ë¸ í•˜ì´í¼íŒŒë¼ë¯¸í„° ---
    'EMBEDDING_DIM': 128,
    'HIDDEN_DIM': 128,
    'N_LAYERS': 1,
    'DROPOUT': 0.6,
    
    # --- í•™ìŠµ í•˜ì´í¼íŒŒë¼ë¯¸í„° ---
    'EPOCHS': 20,
    'LEARNING_RATE': 0.001,
    'BATCH_SIZE': 128,
    'MAX_LEN': 40,                   # ë¬¸ì¥ ìµœëŒ€ ê¸¸ì´
    'PATIENCE': 3,                   # ì¡°ê¸° ì¢…ë£Œ ì¡°ê±´
    'WEIGHT_DECAY': 1e-5,            # ê°€ì¤‘ì¹˜ ê°ì‡  (L2 ê·œì œ)
    
    # --- ê¸°íƒ€ ì„¤ì • ---
    'SEED': 42,
    'DEVICE': torch.device('cuda' if torch.cuda.is_available() else 'cpu'),
    'MODEL_SAVE_PATH': 'best_model.pth',
}

# ====================================================================================
# 3. ìœ í‹¸ë¦¬í‹° í•¨ìˆ˜
# ====================================================================================
def seed_everything(seed):
    """ì¬í˜„ì„±ì„ ìœ„í•œ ì‹œë“œ ê³ ì • í•¨ìˆ˜"""
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = True

# ====================================================================================
# 4. ë°ì´í„° ì¤€ë¹„ í•¨ìˆ˜
# ====================================================================================
def load_and_preprocess_data():
    """NSMC ë°ì´í„°ë¥¼ ë¡œë“œí•˜ê³  ì „ì²˜ë¦¬í•˜ëŠ” í•¨ìˆ˜"""
    # data_path = os.path.join(os.getenv("HOME"), 'work', 'workplace', 'AIFFEL_quest_rs', 'Exploration', 'Ex05', 'sentiment_classification', 'data')
    data_path = os.path.join(os.getenv("HOME"), 'work', 'mecab2', 'data')
    train_data = pd.read_table(os.path.join(data_path, 'ratings_train.txt'))
    test_data = pd.read_table(os.path.join(data_path, 'ratings_test.txt'))
    
    # --- í›ˆë ¨ ë°ì´í„° ì •ì œ ---
    train_data.dropna(subset=['document'], inplace=True)
    train_data.drop_duplicates(subset=['document'], inplace=True)
    
    # --- í…ŒìŠ¤íŠ¸ ë°ì´í„° ì •ì œ ---
    test_data.dropna(subset=['document'], inplace=True)
    test_data.drop_duplicates(subset=['document'], inplace=True)
    
    # --- ë°ì´í„°ì…‹ ë¶„ë¦¬ ---
    train_set, val_set = train_test_split(train_data, test_size=0.2, random_state=CFG['SEED'], stratify=train_data['label'])
    
    print("âœ… ë°ì´í„° ë¡œë“œ ë° ì „ì²˜ë¦¬ ì™„ë£Œ")
    return train_set, val_set, test_data

# ====================================================================================
# 5. í† í¬ë‚˜ì´ì € ìƒì„± í•¨ìˆ˜ (KoNLPy ë¡œì§ ìˆ˜ì •)
# ====================================================================================
def get_tokenizer(cfg, train_df):
    """CFGì— ë”°ë¼ SentencePiece ë˜ëŠ” KoNLPy í† í¬ë‚˜ì´ì €ì™€ vocab_sizeë¥¼ ë°˜í™˜"""
    
    if cfg['TOKENIZER_TYPE'] == 'SentencePiece':
        # --- (SentencePiece ë¡œì§ì€ ì´ì „ê³¼ ë™ì¼) ---
        corpus_path = 'nsmc_corpus.txt'
        model_prefix = f'nsmc_{cfg["SP_MODEL_TYPE"]}_{cfg["SP_VOCAB_SIZE"]}'
        train_df['document'].to_csv(corpus_path, index=False, header=False)
        spm.SentencePieceTrainer.train(
            f'--input={corpus_path} --model_prefix={model_prefix} '
            f'--vocab_size={cfg["SP_VOCAB_SIZE"]} --model_type={cfg["SP_MODEL_TYPE"]}'
        )
        processor = spm.SentencePieceProcessor()
        processor.load(f'{model_prefix}.model')
        vocab_size = processor.get_piece_size()

        def tokenize_fn(corpus, max_len):
            sequences = []
            for sentence in corpus:
                ids = processor.encode_as_ids(str(sentence))
                ids = ids[:max_len] if len(ids) > max_len else ids + [0] * (max_len - len(ids))
                sequences.append(ids)
            return torch.tensor(sequences, dtype=torch.long)
            
        print(f"âœ… SentencePiece í† í¬ë‚˜ì´ì € ì¤€ë¹„ ì™„ë£Œ (vocab_size: {vocab_size})")
        return tokenize_fn, vocab_size

    elif cfg['TOKENIZER_TYPE'] == 'KoNLPy':
        # --- KoNLPy í† í¬ë‚˜ì´ì € ìƒì„± (vocab_size ì œí•œ ë¡œì§ ì¶”ê°€) ---
        if cfg['KONLPY_TOKENIZER'] == 'Mecab':
            tokenizer = Mecab()
        else:
            from konlpy.tag import Okt
            tokenizer = Okt()
            
        # 1. ë‹¨ì–´ ë¹ˆë„ìˆ˜ ê³„ì‚°
        word_counts = {}
        for sentence in tqdm(train_df['document'], desc="KoNLPy Freq. Counting"):
            tokens = tokenizer.morphs(str(sentence))
            for token in tokens:
                word_counts[token] = word_counts.get(token, 0) + 1
        
        # 2. ë¹ˆë„ìˆ˜ ê¸°ì¤€ìœ¼ë¡œ ìƒìœ„ ë‹¨ì–´ ì„ íƒ
        # <PAD>, <UNK> í† í°ì„ ìœ„í•´ (vocab_size - 2)ê°œë§Œ ì„ íƒ
        sorted_words = sorted(word_counts, key=word_counts.get, reverse=True)
        top_words = sorted_words[:cfg['KONLPY_VOCAB_SIZE'] - 2]
        
        # 3. ìµœì¢… ë‹¨ì–´ ì‚¬ì „ êµ¬ì¶•
        word_index = {'<PAD>': 0, '<UNK>': 1}
        for word in top_words:
            word_index[word] = len(word_index)
            
        vocab_size = len(word_index)

        def tokenize_fn(corpus, max_len):
            sequences = []
            for sentence in corpus:
                tokens = tokenizer.morphs(str(sentence))
                ids = [word_index.get(token, 1) for token in tokens] # ì‚¬ì „ì— ì—†ìœ¼ë©´ <UNK> (1)
                ids = ids[:max_len] if len(ids) > max_len else ids + [0] * (max_len - len(ids))
                sequences.append(ids)
            return torch.tensor(sequences, dtype=torch.long)
            
        print(f"âœ… KoNLPy({cfg['KONLPY_TOKENIZER']}) í† í¬ë‚˜ì´ì € ì¤€ë¹„ ì™„ë£Œ (vocab_size: {vocab_size})")
        return tokenize_fn, vocab_size

# ====================================================================================
# 6. ëª¨ë¸ ì •ì˜
# ====================================================================================
class SentimentGRU(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, n_layers, dropout):
        super(SentimentGRU, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.gru = nn.GRU(embedding_dim, hidden_dim, num_layers=n_layers, batch_first=True)
        self.dropout = nn.Dropout(dropout)
        self.fc = nn.Linear(hidden_dim, 1)

    def forward(self, x):
        embedded = self.embedding(x)
        gru_out, hidden = self.gru(embedded)
        last_output = gru_out[:, -1, :]
        last_output = self.dropout(last_output)
        output = self.fc(last_output)
        return output

# ====================================================================================
# 7. í•™ìŠµ ë° í‰ê°€ í•¨ìˆ˜ (ì¶œë ¥ê°’ ìˆ˜ì •)
# ====================================================================================
def run_experiment(cfg, train_set, val_set, test_set):
    """í•˜ë‚˜ì˜ ì„¤ì •(CFG)ìœ¼ë¡œ ì „ì²´ ì‹¤í—˜ì„ ì‹¤í–‰í•˜ëŠ” ë©”ì¸ í•¨ìˆ˜"""
    
    # --- 1. í† í¬ë‚˜ì´ì € ë° ë°ì´í„°ë¡œë” ì¤€ë¹„ ---
    tokenize_fn, vocab_size = get_tokenizer(cfg, train_set)
    
    X_train = tokenize_fn(train_set['document'].tolist(), cfg['MAX_LEN'])
    y_train = torch.tensor(train_set['label'].values, dtype=torch.float32)
    X_val = tokenize_fn(val_set['document'].tolist(), cfg['MAX_LEN'])
    y_val = torch.tensor(val_set['label'].values, dtype=torch.float32)
    X_test = tokenize_fn(test_set['document'].tolist(), cfg['MAX_LEN'])
    y_test = torch.tensor(test_set['label'].values, dtype=torch.float32)

    train_loader = DataLoader(TensorDataset(X_train, y_train), batch_size=cfg['BATCH_SIZE'], shuffle=True)
    val_loader = DataLoader(TensorDataset(X_val, y_val), batch_size=cfg['BATCH_SIZE'], shuffle=False)
    test_loader = DataLoader(TensorDataset(X_test, y_test), batch_size=cfg['BATCH_SIZE'], shuffle=False)
    
    # --- 2. ëª¨ë¸, ì†ì‹¤í•¨ìˆ˜, ì˜µí‹°ë§ˆì´ì € ì •ì˜ ---
    model = SentimentGRU(vocab_size, cfg['EMBEDDING_DIM'], cfg['HIDDEN_DIM'], cfg['N_LAYERS'], cfg['DROPOUT']).to(cfg['DEVICE'])
    loss_fn = nn.BCEWithLogitsLoss()
    optimizer = optim.Adam(model.parameters(), lr=cfg['LEARNING_RATE'], weight_decay=cfg['WEIGHT_DECAY'])
    
    # --- 3. í•™ìŠµ ë° ì¡°ê¸° ì¢…ë£Œ ---
    patience_counter = 0
    best_loss = np.Inf
    
    print("\nğŸš€ ëª¨ë¸ í•™ìŠµì„ ì‹œì‘í•©ë‹ˆë‹¤...")
    for epoch in range(cfg['EPOCHS']):
        # --- í›ˆë ¨ ë‹¨ê³„ ---
        model.train()
        train_loss, train_correct, train_total = 0, 0, 0
        for inputs, labels in tqdm(train_loader, desc=f"Epoch {epoch+1:02d} [Train]"):
            inputs, labels = inputs.to(cfg['DEVICE']), labels.to(cfg['DEVICE'])
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = loss_fn(outputs.squeeze(), labels)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()
            preds = torch.sigmoid(outputs.squeeze()) > 0.5
            train_correct += (preds == labels).sum().item()
            train_total += labels.size(0)

        # <<<--- í›ˆë ¨ ê²°ê³¼ ê³„ì‚° --- START ---
        # ë§¤ ì—í¬í¬ì˜ í›ˆë ¨ì´ ëë‚˜ë©´ í‰ê·  ì†ì‹¤ê³¼ ì •í™•ë„ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.
        avg_train_loss = train_loss / len(train_loader)
        train_accuracy = train_correct / train_total
        # <<<--- í›ˆë ¨ ê²°ê³¼ ê³„ì‚° --- END ---
        
        # --- ê²€ì¦ ë‹¨ê³„ ---
        model.eval()
        val_loss, val_correct, val_total = 0, 0, 0
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(cfg['DEVICE']), labels.to(cfg['DEVICE'])
                outputs = model(inputs)
                loss = loss_fn(outputs.squeeze(), labels)
                val_loss += loss.item()
                preds = torch.sigmoid(outputs.squeeze()) > 0.5
                val_correct += (preds == labels).sum().item()
                val_total += labels.size(0)

        avg_val_loss = val_loss / len(val_loader)
        val_accuracy = val_correct / val_total

        # <<<--- ì¶œë ¥ë¬¸ ìˆ˜ì • --- START ---
        # ê¸°ì¡´ ì¶œë ¥ë¬¸ì— avg_train_lossì™€ train_accuracyë¥¼ ì¶”ê°€í•©ë‹ˆë‹¤.
        print(f"Epoch {epoch+1:02d} | "
              f"Train Loss: {avg_train_loss:.4f} | Train Acc: {train_accuracy*100:.2f}% | "
              f"Val Loss: {avg_val_loss:.4f} | Val Acc: {val_accuracy*100:.2f}%")
        # <<<--- ì¶œë ¥ë¬¸ ìˆ˜ì • --- END ---
        
        if avg_val_loss < best_loss:
            best_loss = avg_val_loss
            torch.save(model.state_dict(), cfg['MODEL_SAVE_PATH'])
            patience_counter = 0
        else:
            patience_counter += 1
        
        if patience_counter >= cfg['PATIENCE']:
            print(f"ğŸ›‘ Early stopping triggered after {epoch+1} epochs.")
            break
            
    # --- 4. ìµœì¢… í‰ê°€ ---
    print(f"\nğŸ§ª ìµœê³  ì„±ëŠ¥ ëª¨ë¸('{cfg['MODEL_SAVE_PATH']}')ë¡œ ìµœì¢… í‰ê°€ë¥¼ ì‹œì‘í•©ë‹ˆë‹¤...")
    model.load_state_dict(torch.load(cfg['MODEL_SAVE_PATH']))
    model.eval()
    test_loss, test_correct, test_total = 0, 0, 0
    with torch.no_grad():
        for inputs, labels in tqdm(test_loader, desc="[Testing]"):
            inputs, labels = inputs.to(cfg['DEVICE']), labels.to(cfg['DEVICE'])
            outputs = model(inputs)
            loss = loss_fn(outputs.squeeze(), labels)
            test_loss += loss.item()
            preds = torch.sigmoid(outputs.squeeze()) > 0.5
            test_correct += (preds == labels).sum().item()
            test_total += labels.size(0)

    avg_test_loss = test_loss / len(test_loader)
    test_accuracy = test_correct / test_total
    
    print("\nğŸ‰ ëª¨ë¸ í‰ê°€ê°€ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.")
    print("-" * 40)
    print(f"  - ìµœì¢… í…ŒìŠ¤íŠ¸ ì†ì‹¤ (Loss): {avg_test_loss:.4f}")
    print(f"  - ìµœì¢… í…ŒìŠ¤íŠ¸ ì •í™•ë„ (Accuracy): {test_accuracy*100:.2f}%")
    print("-" * 40)
    
    return {'loss': avg_test_loss, 'accuracy': test_accuracy}

# ====================================================================================
# 8. ë©”ì¸ ì‹¤í–‰ ë¸”ë¡
# ====================================================================================
if __name__ == '__main__':
    seed_everything(CFG['SEED'])
    train_set, val_set, test_set = load_and_preprocess_data()
    
    # --- ì—¬ê¸°ì„œë¶€í„° ì‹¤í—˜ì„ ì‹¤í–‰í•©ë‹ˆë‹¤ ---
    # ì˜ˆì‹œ 1: SentencePiece (bpe, vocab_size=10000) ì‹¤í—˜
    CFG['TOKENIZER_TYPE'] = 'SentencePiece'
    CFG['SP_MODEL_TYPE'] = 'bpm'
    # results = run_experiment(CFG, train_set, val_set, test_set)
    
    # ì˜ˆì‹œ 2: KoNLPy (Mecab) ì‹¤í—˜ (CFG ë³€ê²½ í›„ ì‹¤í–‰)
    # CFG['TOKENIZER_TYPE'] = 'KoNLPy'
    # CFG['KONLPY_TOKENIZER'] = 'Mecab'
    # results_mecab = run_experiment(CFG, train_set, val_set, test_set)
    
    # ì˜ˆì‹œ 3: SentencePiece (unigram, vocab_size=16000) ì‹¤í—˜ (CFG ë³€ê²½ í›„ ì‹¤í–‰)
    # CFG['TOKENIZER_TYPE'] = 'SentencePiece'
    # CFG['SP_MODEL_TYPE'] = 'unigram'
    # CFG['SP_VOCAB_SIZE'] = 16000
    # results_sp_uni = run_experiment(CFG, train_set, val_set, test_set)
```

```python
# sp_tokenize í•¨ìˆ˜ êµ¬í˜„ ë° í…ŒìŠ¤íŠ¸
import sentencepiece as spm
import torch

# 1. í•™ìŠµëœ SentencePiece ëª¨ë¸ ë¡œë“œ
sp = spm.SentencePieceProcessor()
# ì´ì „ì— í•™ìŠµì‹œí‚¨ ëª¨ë¸ íŒŒì¼ëª…ì„ ì‚¬ìš©í•©ë‹ˆë‹¤.
sp.load('nsmc_bpe_model.model')

# 2. sp_tokenize í•¨ìˆ˜ ì •ì˜
def sp_tokenize(processor, corpus, max_len):
    """
    SentencePiece ëª¨ë¸ì„ ì‚¬ìš©í•˜ì—¬ ë¬¸ì¥ ë¦¬ìŠ¤íŠ¸ë¥¼ í† í°í™”í•˜ê³  íŒ¨ë”©ì„ ì ìš©í•©ë‹ˆë‹¤.
    
    Args:
        processor (SentencePieceProcessor): ë¡œë“œëœ SentencePiece ëª¨ë¸ ê°ì²´
        corpus (list of str): í† í°í™”í•  ë¬¸ì¥ë“¤ì˜ ë¦¬ìŠ¤íŠ¸
        max_len (int): ëª¨ë“  ì‹œí€€ìŠ¤ì˜ ê¸¸ì´ë¥¼ ë§ì¶œ ìµœëŒ€ ê¸¸ì´
    
    Returns:
        torch.Tensor: í† í°í™” ë° íŒ¨ë”©ì´ ì™„ë£Œëœ íŒŒì´í† ì¹˜ í…ì„œ
    """
    tokenized_sequences = []
    for sentence in corpus:
        # ê° ë¬¸ì¥ì„ ì •ìˆ˜ ID ì‹œí€€ìŠ¤ë¡œ ë³€í™˜
        ids = processor.encode_as_ids(sentence)
        
        # íŒ¨ë”© ë˜ëŠ” ì˜ë¼ë‚´ê¸°
        if len(ids) > max_len:
            # ìµœëŒ€ ê¸¸ì´ë¥¼ ì´ˆê³¼í•˜ë©´ ì˜ë¼ëƒ„
            ids = ids[:max_len]
        else:
            # ìµœëŒ€ ê¸¸ì´ë³´ë‹¤ ì§§ìœ¼ë©´ 0ìœ¼ë¡œ íŒ¨ë”© ì¶”ê°€
            ids += [0] * (max_len - len(ids))
            
        tokenized_sequences.append(ids)
    
    # ë¦¬ìŠ¤íŠ¸ë¥¼ íŒŒì´í† ì¹˜ í…ì„œë¡œ ë³€í™˜
    return torch.tensor(tokenized_sequences, dtype=torch.long)

# 3. ìƒ˜í”Œ ë¬¸ì¥ìœ¼ë¡œ ë™ì‘ í™•ì¸
sample_corpus = [
    "ì´ ì˜í™” ì •ë§ ìµœê³ ì˜ˆìš”!",
    "ì •ë§ ì¬ë¯¸ì—†ë‹¤. ë¹„ì¶”.",
    "ë°°ìš°ë“¤ ì—°ê¸°ë ¥ì´ ì•„ê¹Œìš´ ì˜í™”."
]
max_sequence_length = 15  # íŒ¨ë”©ì„ ì ìš©í•  ìµœëŒ€ ë¬¸ì¥ ê¸¸ì´

# í•¨ìˆ˜ í˜¸ì¶œ
token_tensor = sp_tokenize(sp, sample_corpus, max_len=max_sequence_length)

# 4. ê²°ê³¼ ì¶œë ¥
print("âœ… sp_tokenize í•¨ìˆ˜ ë™ì‘ í™•ì¸")
print("-" * 30)
for i, sentence in enumerate(sample_corpus):
    print(f"ì›ë³¸ ë¬¸ì¥ {i+1}: {sentence}")
    print(f"í† í°í™” ê²°ê³¼: {token_tensor[i]}")
    print()

print(f"ìµœì¢… í…ì„œì˜ í¬ê¸°: {token_tensor.shape}")
```

```python
# 1. PyTorch ë¼ì´ë¸ŒëŸ¬ë¦¬ import
import torch
import torch.nn as nn

# 2. ëª¨ë¸ ì„¤ê³„: nn.Moduleì„ ìƒì†ë°›ì•„ SentimentGRU í´ë˜ìŠ¤ ì •ì˜
class SentimentGRU(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, n_layers, dropout):
        """
        ëª¨ë¸ì˜ ë ˆì´ì–´ë¥¼ ì •ì˜í•˜ëŠ” ìƒì„±ì
        
        Args:
            vocab_size (int): ë‹¨ì–´ ì§‘í•©ì˜ í¬ê¸° (SentencePiece ëª¨ë¸ì˜ vocab_sizeì™€ ì¼ì¹˜)
            embedding_dim (int): ê° ë‹¨ì–´ë¥¼ í‘œí˜„í•  ì„ë² ë”© ë²¡í„°ì˜ ì°¨ì›
            hidden_dim (int): GRU ë ˆì´ì–´ì˜ ì€ë‹‰ ìƒíƒœ(hidden state) ì°¨ì›
            n_layers (int): ìŒ“ì„ GRU ë ˆì´ì–´ì˜ ê°œìˆ˜
            dropout (float): ë“œë¡­ì•„ì›ƒ ë¹„ìœ¨
        """
        super(SentimentGRU, self).__init__()
        
        # --- ë ˆì´ì–´ ì •ì˜ ---
        # 1. ì„ë² ë”© ë ˆì´ì–´: ì •ìˆ˜ ì¸ë±ìŠ¤ -> ë°€ì§‘ ë²¡í„°
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        
        # 2. GRU ë ˆì´ì–´: ì„ë² ë”© ë²¡í„° ì‹œí€€ìŠ¤ë¥¼ ì…ë ¥ë°›ì•„ ë¬¸ë§¥ ì •ë³´ ì¶”ì¶œ
        # batch_first=True: ì…ë ¥ í…ì„œì˜ ì°¨ì›ì„ (batch_size, seq_len, embedding_dim)ìœ¼ë¡œ ì„¤ì •
        self.gru = nn.GRU(embedding_dim, 
                          hidden_dim, 
                          num_layers=n_layers, 
                          batch_first=True, 
                          dropout=dropout)
        
        # 3. ì™„ì „ ì—°ê²° ë ˆì´ì–´ (Dense Layer): GRUì˜ ìµœì¢… ì¶œë ¥ì„ ë°›ì•„ ê¸ì •/ë¶€ì •ì„ ì˜ˆì¸¡
        self.fc = nn.Linear(hidden_dim, 1)

    def forward(self, x):
        """
        ëª¨ë¸ì˜ ìˆœì „íŒŒ ë¡œì§ì„ ì •ì˜
        ë°ì´í„°ê°€ ê° ë ˆì´ì–´ë¥¼ ê±°ì¹˜ëŠ” ìˆœì„œë¥¼ ê²°ì •í•©ë‹ˆë‹¤.
        
        Args:
            x (torch.Tensor): ì…ë ¥ ë°ì´í„° (í† í°í™” ë° íŒ¨ë”©ëœ ì •ìˆ˜ ì‹œí€€ìŠ¤ í…ì„œ)
        
        Returns:
            torch.Tensor: ëª¨ë¸ì˜ ìµœì¢… ì˜ˆì¸¡ ê²°ê³¼ (logit)
        """
        # --- ë°ì´í„° íë¦„ ì •ì˜ ---
        # x shape: (batch_size, max_len)
        
        # 1. ì„ë² ë”© ë ˆì´ì–´ í†µê³¼
        embedded = self.embedding(x)
        # embedded shape: (batch_size, max_len, embedding_dim)
        
        # 2. GRU ë ˆì´ì–´ í†µê³¼
        # gru_out: ëª¨ë“  ì‹œì (time step)ì˜ ì€ë‹‰ ìƒíƒœ
        # hidden: ë§ˆì§€ë§‰ ì‹œì ì˜ ì€ë‹‰ ìƒíƒœ
        gru_out, hidden = self.gru(embedded)
        # gru_out shape: (batch_size, max_len, hidden_dim)
        # hidden shape: (n_layers, batch_size, hidden_dim)
        
        # 3. ë§ˆì§€ë§‰ ì‹œì ì˜ ì¶œë ¥ë§Œ ì„ íƒ
        # ë¬¸ì¥ ì „ì²´ì˜ ì˜ë¯¸ë¥¼ ë‹´ê³  ìˆëŠ” ë§ˆì§€ë§‰ ë‹¨ì–´ì˜ ì¶œë ¥ì„ ì‚¬ìš©
        last_output = gru_out[:, -1, :]
        # last_output shape: (batch_size, hidden_dim)
        
        # 4. ì™„ì „ ì—°ê²° ë ˆì´ì–´ í†µê³¼
        output = self.fc(last_output)
        # output shape: (batch_size, 1)
        
        return output

# 3. ëª¨ë¸ ì¸ìŠ¤í„´ìŠ¤ ìƒì„± ë° êµ¬ì¡° í™•ì¸
# --- í•˜ì´í¼íŒŒë¼ë¯¸í„° ì •ì˜ ---
VOCAB_SIZE = 10000       # SentencePiece ëª¨ë¸ì˜ vocab_sizeì™€ ë™ì¼í•´ì•¼ í•¨
EMBEDDING_DIM = 128      # ì„ë² ë”© ë²¡í„° ì°¨ì›
HIDDEN_DIM = 128         # GRU ì€ë‹‰ ìƒíƒœ ì°¨ì›
N_LAYERS = 2             # GRU ë ˆì´ì–´ ê°œìˆ˜
DROPOUT = 0.6            # ë“œë¡­ì•„ì›ƒ ë¹„ìœ¨

# ëª¨ë¸ ê°ì²´ ìƒì„±
model = SentimentGRU(VOCAB_SIZE, EMBEDDING_DIM, HIDDEN_DIM, N_LAYERS, DROPOUT)

# ëª¨ë¸ êµ¬ì¡° ì¶œë ¥
print("âœ… ëª¨ë¸ ì„¤ê³„ê°€ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.")
print(model)
```

```python
# 1. í•„ìš” ë¼ì´ë¸ŒëŸ¬ë¦¬ import
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from tqdm.notebook import tqdm
import numpy as np # best_loss ì´ˆê¸°í™”ë¥¼ ìœ„í•´ import

# ====================================================================================
# 1. ë°ì´í„° ì¤€ë¹„ ë° DataLoader ìƒì„± (ì´ì „ê³¼ ë™ì¼)
# ====================================================================================

# 1-1. í•˜ì´í¼íŒŒë¼ë¯¸í„° ì •ì˜
MAX_LEN = 40
BATCH_SIZE = 128

# 1-2. ë°ì´í„°ì…‹ì„ í† í°í™”í•˜ê³  í…ì„œë¡œ ë³€í™˜
X_train_tensor = sp_tokenize(sp, train_set['document'].tolist(), MAX_LEN)
y_train_tensor = torch.tensor(train_set['label'].values, dtype=torch.float32)

X_val_tensor = sp_tokenize(sp, val_set['document'].tolist(), MAX_LEN)
y_val_tensor = torch.tensor(val_set['label'].values, dtype=torch.float32)

# 1-3. DataLoader ìƒì„±
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)

val_dataset = TensorDataset(X_val_tensor, y_val_tensor)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)

print("âœ… ë°ì´í„°ë¡œë” ìƒì„±ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.")

# ====================================================================================
# 2. ëª¨ë¸ í•™ìŠµ ë° í‰ê°€ ì‹¤í–‰ (ì¡°ê¸° ì¢…ë£Œ ë¡œì§ ì¶”ê°€)
# ====================================================================================

# 2-1. í•™ìŠµ ê´€ë ¨ í•˜ì´í¼íŒŒë¼ë¯¸í„° ë° ì„¤ì •
EPOCHS = 20 # ìµœëŒ€ í•™ìŠµ íšŸìˆ˜ë¥¼ ë„‰ë„‰í•˜ê²Œ ì„¤ì •
LEARNING_RATE = 0.001

# --- ì¡°ê¸° ì¢…ë£Œ ê´€ë ¨ ì„¤ì • ---
patience = 3                 # ì„±ëŠ¥ ê°œì„ ì„ ê¸°ë‹¤ë¦´ íšŸìˆ˜
patience_counter = 0         # ì„±ëŠ¥ ë¯¸ê°œì„  íšŸìˆ˜ ì¹´ìš´í„°
best_loss = np.Inf           # ì´ˆê¸° ìµœì € ì†ì‹¤ê°’ì„ ë¬´í•œëŒ€(inf)ë¡œ ì„¤ì •
best_model_path = 'best_model.pth' # ìµœê³  ì„±ëŠ¥ ëª¨ë¸ì´ ì €ì¥ë  ê²½ë¡œ

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

loss_fn = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE, weight_decay=1e-5)

print(f"ì‚¬ìš© ë””ë°”ì´ìŠ¤: {device}")
print("\nğŸš€ ëª¨ë¸ í•™ìŠµì„ ì‹œì‘í•©ë‹ˆë‹¤ (ì¡°ê¸° ì¢…ë£Œ ì ìš©)...")

# 2-2. í•™ìŠµ ë£¨í”„ ì‹¤í–‰
for epoch in range(EPOCHS):
    # --- í›ˆë ¨ ë‹¨ê³„ (ì´ì „ê³¼ ë™ì¼) ---
    model.train()
    train_loss, train_correct, train_total = 0, 0, 0
    
    for inputs, labels in tqdm(train_loader, desc=f"Epoch {epoch+1:02d}/{EPOCHS} [Training]"):
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = loss_fn(outputs.squeeze(), labels)
        loss.backward()
        optimizer.step()
        
        train_loss += loss.item()
        preds = torch.sigmoid(outputs.squeeze()) > 0.5
        train_correct += (preds == labels).sum().item()
        train_total += labels.size(0)

    avg_train_loss = train_loss / len(train_loader)
    train_accuracy = train_correct / train_total

    # --- ê²€ì¦ ë‹¨ê³„ (ì´ì „ê³¼ ë™ì¼) ---
    model.eval()
    val_loss, val_correct, val_total = 0, 0, 0
    
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = loss_fn(outputs.squeeze(), labels)
            val_loss += loss.item()
            preds = torch.sigmoid(outputs.squeeze()) > 0.5
            val_correct += (preds == labels).sum().item()
            val_total += labels.size(0)

    avg_val_loss = val_loss / len(val_loader)
    val_accuracy = val_correct / val_total
    
    print(f"Epoch {epoch+1:02d}/{EPOCHS} | "
          f"Train Loss: {avg_train_loss:.4f} | Train Acc: {train_accuracy*100:.2f}% | "
          f"Val Loss: {avg_val_loss:.4f} | Val Acc: {val_accuracy*100:.2f}%")

    # --- ì¡°ê¸° ì¢…ë£Œ ë¡œì§ ---
    if avg_val_loss < best_loss:
        # ê²€ì¦ ì†ì‹¤ì´ ê°œì„ ë˜ë©´, best_lossë¥¼ ì—…ë°ì´íŠ¸í•˜ê³  ëª¨ë¸ ê°€ì¤‘ì¹˜ë¥¼ ì €ì¥
        best_loss = avg_val_loss
        torch.save(model.state_dict(), best_model_path)
        patience_counter = 0 # ì¹´ìš´í„° ì´ˆê¸°í™”
        print(f"ğŸ’¡ Validation loss improved. Saving model to '{best_model_path}'")
    else:
        # ê²€ì¦ ì†ì‹¤ì´ ê°œì„ ë˜ì§€ ì•Šìœ¼ë©´ ì¹´ìš´í„° ì¦ê°€
        patience_counter += 1
        print(f"âš ï¸ Validation loss did not improve. Patience: {patience_counter}/{patience}")

    if patience_counter >= patience:
        # ì¹´ìš´í„°ê°€ ì§€ì •ëœ íšŸìˆ˜ì— ë„ë‹¬í•˜ë©´ í•™ìŠµ ì¤‘ë‹¨
        print(f"ğŸ›‘ Early stopping triggered after {epoch+1} epochs.")
        break

# --- í•™ìŠµ ì¢…ë£Œ í›„ ìµœê³  ì„±ëŠ¥ ëª¨ë¸ ë¡œë“œ ---
print(f"\nğŸ‰ ëª¨ë¸ í•™ìŠµì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤. ìµœê³  ì„±ëŠ¥ì˜ ëª¨ë¸('{best_model_path}')ì„ ë¡œë“œí•©ë‹ˆë‹¤.")
model.load_state_dict(torch.load(best_model_path))
```
